```python
import xarray as xr
xr.set_options(keep_attrs=True)
```

```python
dimension="YC"; typ='float32'; engine='h5netcdf'

if dimension == 'XC':
    chunk_d = {'time': 1, 'Z': 52, 'YC': 588, 'XC': 2160}
elif dimension == 'YC':
     chunk_d = {'time': 1, 'Z': 52, 'YC': 588, 'XC': 2160}

density_ds = xr.open_mfdataset('nc/density.nc',
                               #engine=engine,
                               #decode_cf=False,
                               chunks=chunk_d,
                               combine='by_coords',
                               data_vars='minimal',
                               coords='minimal',
                               compat='override',
                               parallel=True
                               ).astype(typ)
```

```python
density_ds
```

```python
density_ds.Density.differentiate(dimension).to_dataset().rename_vars({'Density': 'Density_Gradient'})
```

```python
density_ds['Density'].attrs['units'] = 'kg m-3 box-1'
```

```python
#density_ds['Density'].attrs['standard_name'] = 'Density_Gradient_XC'
density_ds['Density'].attrs['long_name'] = 'Density Gradient XC'
```
